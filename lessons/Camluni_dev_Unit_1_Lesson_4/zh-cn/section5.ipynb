## 学习
我们先来复习一下返回值:

1、python函数使用return语句返回 "返回值"，可以将其赋给其它变量作其它的用处

2、所有函数都有返回值，如果没有return语句，会隐式地调用 return None 作为返回值

3、一个函数可以存在多条return语句，但只有一条可以被执行，如果没有一条reutrn语句被执行，同样会隐式调用return None作为返回值

4、如果函数执行了return语句，函数会立刻返回，结束调用，return之后的其它语句都不会被执行了(可用于结束代码块)

返回值丰富了函数的功能，给我们的编程带来了极大的遍历！但是如果我在一个函数中，想要返回多个数据怎么办呢？这个时候我们就需要用到函数的“返回多个值”的功能了！

基本语法如下：

```python
def 函数名(形式参数):
    ...
    return 返回值1 , 返回值2 , 返回值3 ...
    
```

我们可以看到,如果需要多个返回值的话，只需要在每个返回值之前使用逗号隔开即可！函数返回多个值后，当我们调用函数的时候，返回值会以Tuple的形式返回给函数，所以，如果我们使用变量接收这个函数的话，那这个变量就是一个Tuple数据类型的变量。每个返回值会以元素的形形式按照顺序保存到Tuple中。

````python
# section5
def agent_up(step = 1):
    for i in range(step):
        agent.move('up')
def agent_right(step = 1):
    for i in range(step):
        agent.move('right')
def agent_left(step = 1):
    for i in range(step):
        agent.move('left')
def get_agent_front_pos(distance):
    facing = agent.rotation
    step = (0, 0, 0)
    if facing == 0:
        step = (0, 0, distance)
    elif facing == 90:
        step = (-distance, 0, 0)
    elif facing == -90:
        step = (-distance, 0, 0)
    else:
        step =  (0, 0, distance)
    return agent.position + step
def set_block_by_xyz(material,*pos):
    say(material+'方块放置成功！')
    if world.is_block(pos,('concrete',4)):
        world.set((175,104,261),'air')
    world.set(pos,material)

def get_block_pos_by_id(block_id):
    blocks = []
    for i in range(16):
        blocks.append(('concrete',i))
    for i in range(6):
        for i in range(7):
            for block in blocks:
                pos = get_agent_front_pos(2)
                if world.is_block(pos,block):
                    x,y,z= pos
                    bid = block[1]
                    if bid == block_id:

                        #在前面的程序中，我们已经对方块进行了比对，现在得到了这个方块位置的x,y,z坐标
                         #我们需要一次性返回x,y,z这三个值
#coding                       
                        return x,y,z
                    break
                    
            agent_right()
        agent_left(7)
        agent_up()
#coding
# 将调用get_block_pos_by_id的函数返回值传给x,y,z
x,y,z = get_block_pos_by_id(4)
# 调用set_block函数，传递'air'方块和x,y,z数字
set_block_by_xyz('air',x,y,z)

````
