## 知识点

- 函数返回表达式


## 学习
请查看下面的例子：
```python
def num_add_num(num_1,num_2):
    result = num_1 + num_2
    return result
```
这是一个非常简单的两个数字相加的函数，result是一个变量，用来接收两个参数相加的结果，然后将这个变量返回给函数。

我们有没有考虑过一个问题，如果我不想使用那么多变量，想把得到的结果直接返回给函数可行吗？答案是可以的，在Python编程中，我们完全可以将一个表达式直接通过返回值的形式返回给函数。例如上面的程序我们可以改造成下面的形式:

```python
def num_add_num(num_1,num_2):
    return num_1 + num_2
```
很显然，直接返回表达式会避免使用一次变量。在某些时候，我们可以使用这种方式提高编程的效率或者减少程序的代码量。

## 任务6-1
小精灵前方的黑板后面存在一些方块，只有一个方块的数据值是4，我们需要找到这个数据值为4的方块的位置，并打碎这个方块前方的黑板方块即可通关！

````python
# section6
def agent_up(step = 1):
    for i in range(step):
        agent.move('up')
def agent_right(step = 1):
    for i in range(step):
        agent.move('right')
def agent_left(step = 1):
    for i in range(step):
        agent.move('left')
def get_agent_front_pos(distence):
    facing = agent.rotation
    step = (0, 0, 0)
    if facing == 0:
        step = (0, 0, distence)
    elif facing == 90:
        step = (-distence, 0, 0)
    elif facing == -90:
        step = (-distence, 0, 0)
    else:
        step =  (0, 0, distence)
    return agent.position + step
def set_block_by_xyz(material,*pos):
    say(material+'方块放置成功！')
    if world.is_block(pos,('concrete',4)):
        world.set((175,104,261),'air')
    world.set(pos,material)

def get_block_data_by_id(block_id):
    blocks = []
    for i in range(15):
        blocks.append(('concrete',i))
    for i in range(3):
        for i in range(3):
            for block in blocks:
                pos = get_agent_front_pos(2)
                if world.is_block(pos,block):
                    bname,bid = block[0],block[1]
                    if bid == block_id:
#coding
                        #在前面的程序中，我们已经对方块进行了比对
                        #现在得到了这个方块位置的name和id值
                        #我们需要返回bname 和 bid 拼接之后的值
                        return bname + str(bid)
                    break
                    
            agent_right()
        agent_left(3)
        agent_up()

# 将调用get_block_data_by_id的函数返回值传给data变量
data = get_block_data_by_id(4)
# 打印data变量并打碎黑板
say(data)
agent.destroy('forward')
````
